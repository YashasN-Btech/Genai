# Hugging Face Pipeline – Sentiment Analysis

```python
from transformers import pipeline

sentiment = pipeline("sentiment-analysis")

texts = [
    "I loved the new movie! The acting was great.",
    "This product was terrible and broke immediately."
]

results = sentiment(texts)

for t, r in zip(texts, results):
    print(t)
    print("→", r)
    print()
```

# Summarization

```python
summarizer = pipeline("summarization")

article = """
Artificial intelligence has made rapid progress in recent years.
Large language models can now perform complex tasks such as reasoning,
translation, and summarization.
"""

summary = summarizer(
    article,
    max_length=60,
    min_length=20,
    do_sample=False
)

print(summary[0]['summary_text'])
```

# Sentence Embeddings (Semantic Similarity)

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
    "How do I cook pasta?",
    "Best way to boil spaghetti",
    "What is the capital of France?"
]

embeddings = model.encode(sentences, convert_to_tensor=True)
print("Embedding shape:", embeddings.shape)
```

```python
query = "How to make spaghetti?"
query_emb = model.encode(query, convert_to_tensor=True)

scores = util.pytorch_cos_sim(query_emb, embeddings)[0]

for s, score in zip(sentences, scores):
    print(f"{score.item():.3f} → {s}")
```

# FastAI – Text Classification (IMDb)

Load dataset

```python
from datasets import load_dataset

ds = load_dataset("imdb")
ds
```

Reduce dataset size (for speed)

```python
small_train = ds['train'].shuffle(seed=42).select(range(4000))
small_test = ds['test'].shuffle(seed=42).select(range(2000))
```

Convert to DataFrame

```python
import pandas as pd

train_df = pd.DataFrame(small_train)
test_df = pd.DataFrame(small_test)
```

Train model

```python
from fastai.text.all import *

dls = TextDataLoaders.from_df(
    train_df,
    text_col='text',
    label_col='label',
    valid_pct=0.1,
    bs=32
)

learn = text_classifier_learner(
    dls,
    AWD_LSTM,
    drop_mult=0.5,
    metrics=accuracy
)
from fastai.callback.progress import ProgressCallback
learn.remove_cb(ProgressCallback)

learn.fine_tune(1)
```
